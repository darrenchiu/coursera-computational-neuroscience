```python
import pickle
with open('c10p1.pickle', 'rb') as f:
    data = pickle.load(f)
    
import numpy as np
```

```python
df = np.array(data['c10p1'])

print(df)
```

```python
# Center-Zero

#create function to center data
center_function = lambda x: x - x.mean()

#apply function to original NumPy array
data_centered = center_function(df)

#view updated Array
print(data_centered)

#display mean of centered array
print(data_centered.mean())
```

```python
import matplotlib.pyplot as plt

# a scatter plot comparing num_children and num_pets
plt.scatter(df[:, 0], df[:, 1], color='red')
plt.show()
```

```python
# w = w + 0.01(vu-v^2w)


w0 = np.array([[0.2], [0.5]])
w1 = np.array([[0.2], [0.5]])


for j in range(0, 10):
    for i in range(0, 100000):

        # pick random row
        u = data_centered[np.random.choice(data_centered.shape[0], size=1), :]

        w0 = w1
        part1 = ((u@w0)[0][0])*u
        part2 = ((u@w0)[0][0])**2 * w0

        w1 = w0 + 0.02*(part1.T - part2)

    print(w1)
```
